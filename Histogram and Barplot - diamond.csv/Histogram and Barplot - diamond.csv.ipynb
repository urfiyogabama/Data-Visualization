```python
import seaborn as sns
import pandas as pd
from numpy import mean,median
import matplotlib.pyplot as plt

df = sns.load_dataset('diamonds')
```

```python
df.head()
```

```python
#Distribution plot from 'carat'
sns.distplot(df['carat'])
```

```python
#'cut' barplot with pandas
count=pd.crosstab(index=df['cut'],columns='count')
count.plot(kind='bar')
```

We can see that the most diamond in data is Ideal, then Premium, Very Good, Good, and Fair.

```python
#cut barplot with seaborn
sns.catplot('cut',data=df,aspect=1.2,kind='count',color='r')
```

```python
#the mean price distribution of different cut
sns.set(style='whitegrid')
ax = sns.barplot(x='cut', y='price', data=df,estimator=mean, ci=68, order=['Ideal','Good','Very Good','Fair','Premium'], color='r')
```

```python
ax=sns.barplot(x='cut',y='price',hue='color',data=df)
ax.legend(loc='upper right',ncol=4)
ax.set_xlabel('Cut',fontdict={'fontsize':15})
ax.set_xlabel('Price',fontdict={'fontsize':15})
```

The price patterns for diamonds of different colors are similar for each cut quality.

```python
#Grouped barplot with right legend and rotated x label
ax=sns.barplot(x='cut',y='price',hue='color',data=df)
ax.legend(loc='upper right',ncol=4)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=13, rotation=20)
```

```python
#Annotate
ideal=df[df['cut']=='Ideal']
x=ideal.index.tolist()[0]
y=len(ideal)
sns.catplot("cut",data=df,aspect=1.2,kind='count',color='b')
plt.annotate('Excellent polish and symmetry ratings;\nreflects almost all the light that enters it',xy=(x,y),xytext=(x+0.3,y+2000),arrowprops=dict(facecolor='red'))
```
